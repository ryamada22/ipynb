```python
from keras.models import Sequential
from keras.layers import Dense

from sklearn.metrics import r2_score
from matplotlib import pyplot
import numpy as np
%matplotlib inline
```

```python
n_sample = 500
n_feature = 5
x = np.random.random(n_sample * n_feature).reshape(n_sample,n_feature)
w = np.random.random(n_feature)
r = np.random.random(n_feature)
y_true = np.zeros(n_sample).reshape(n_sample,1)
for i in np.arange(n_feature):
    y_true[:,0] += x[:,i] ** r[i] * w[i]
    #y_true[:,0] += x[:,i] ** 1 * w[i]
# 第１説明変数と被説明変数との散布図をプロットしてみる
pyplot.scatter(x[:,0],y_true[:,0])
pyplot.show()
```

```python
X1 = x
Y1 = y_true

# create model
model = Sequential()
model.add(Dense(20, activation="tanh", kernel_initializer="uniform", input_dim=5))
model.add(Dense(1, activation="linear", kernel_initializer="uniform"))
```

```python
# Compile model
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
```

```python



# Fit the model
model.fit(X1, Y1, nb_epoch=500, batch_size=10,  verbose=0)

# Calculate predictions
PredTestSet = model.predict(X1)
#PredValSet = model.predict(X2)
```

```python
pyplot.scatter(Y1,PredTestSet)
pyplot.show()
```
